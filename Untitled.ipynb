```python
import numpy as np
import matplotlib.pyplot as plt
  
x = []
y = []
j=0
for line in open('new.txt', 'r'):
    lines = [i for i in line.split()]
    j= j+1
    x.append(j)
    y.append(int(lines[1]))
      
plt.title("Aggregation")
plt.xlabel('Time')
plt.ylabel('Number of nodes that transmited')
plt.yticks(y)
plt.plot(x, y, marker = 'o', c = 'g')
  
plt.show()
```
